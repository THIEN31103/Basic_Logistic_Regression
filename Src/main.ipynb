```python
X = np.array([[1.0, 0.5],
              [2.0, 1.0],
              [3.0, 1.5],
              [2.0, 2.0],
              [1.0, 2.5],
              [2.0 ,3.0],
              [1.0, 3.5],
              [2.0, 4.0]])
Y = np.array([0, 0, 1, 0, 0, 1, 1, 1])
```

```python
X.shape
```

```python
n_samples = X.shape[0]
n_samples
```

```python
X = np.hstack([np.ones((n_samples,1)),X])
X
```

```python
X.shape
```

```python
n_features = X.shape[1]
n_features
```

```python
theta = np.random.rand(n_features)
theta
```

```python
def logistic_function(x):
    return 1/(1+np.exp(-x))

def predict(X,theta):
    z = np.dot(X,theta)
    y_hat = logistic_function(z)
    return z,y_hat

def compute_loss(y_hat, y):
    loss = -1*(y*np.log(y_hat) + (1-y)*np.log(1-y_hat))
    return loss

def compute_gradient(X, y_hat, y):
    gradient = X*(y_hat - y)
    return gradient

def update_weight(gradient, theta, lr):
    theta = theta - lr*gradient
    return theta
```

```python
n_epochs = 10
lr = 0.01
losses = []
for epoch in range(n_epochs):
    epoch_losses = []
    for i in range(n_samples):
        x = X[i,:]
        y = Y[i]

        #y_hat
        z, y_hat = predict(x, theta)

        #loss
        loss = compute_loss(y_hat, y)
        epoch_losses.append(loss)

        #gradient
        gradient = compute_gradient(x, y_hat, y)

        #theta
        theta = update_weight(gradient, theta, lr )
    losses.append(sum(epoch_losses)/len(epoch_losses))
```

```python
def logistic_function(x):
    return 1/(1+np.exp(-x))

def predict(x,theta):
    z = np.dot(x,theta)
    y_hat = logistic_function(z)
    return y_hat

def compute_loss(y_hat, y):
    m = len(y)
    loss = sum(-1*(y*np.log(y_hat) + (1-y)*np.log(1-y_hat)))/m
    return loss

def compute_gradient(X, y_hat, y):
    m = len(y)
    gradient = 1/m*np.dot(x.transpose(), (y_hat - y))
    return gradient

def update_weight(gradient, theta, lr):
    theta = theta - lr*gradient
    return theta
```

```python
#param
n_epochs = 100
lr = 0.001

losses = []
for epoch in range(n_epochs):
    #get all the samples
    x = X
    y = Y

    #predict y_hat
    y_hat = predict(x, theta)

    #compute loss
    loss = compute_loss(y_hat, y)
    losses.append(loss)

    #compute gradient
    gradient = compute_gradient(x, y, y_hat)

    #update weights
    theta = update_weight(gradient, theta, lr)

```
